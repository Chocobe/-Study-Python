```python
import pandas as pd

mpg = pd.read_csv('../sample-data/Doit_Python-main/Data/mpg.csv')

fuel = pd.DataFrame({
    'fl': ['c', 'd', 'e', 'p', 'r'],
    'price_fl': [2.35, 2.38, 2.11, 2.76, 2.22]
})
fuel
```

```python
# 문제 1) mpg 데이터에는 연료 종류를 나타낸 `f1` 변수는 있지만 연료 가격을 나타낸 변수는 없습니다.
# 앞에서 만든 `fuel` 데이터를 이용해 `mpg` 데이터에 `price_f1(연료 가격)` 변수를 추가하세요.
mpg_new = pd.merge(
    left = mpg,
    right = fuel,
    how = 'left',
    on = 'fl'
)
mpg_new
```

```python
# 문제 2) 연료 가격 변수가 잘 추가됐는지 확인하기 위해 `model`, `fl`, `price_fl` 변수를 추출해 앞부분 5행을 출력해 보세요.
mpg_new[['model', 'fl', 'price_fl']].head(5)
```
