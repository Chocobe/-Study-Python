```python
import pandas as pd

mpg = pd.read_csv('../sample-data/Doit_Python-main/Data/mpg.csv');
mpg
```

```python
# 문제 1) mpg 데이터의 `category`는 자동차를 특지에 따라 `suv`, `compact` 등 일곱 종류로 분류한 변수입니다.
# 어떤 차종의 `cty(도시연비)` 가 높은지 비교해 보려고 합니다.
# `category`별 `cty(도시연비)` 평균을 구해보세요.

mpg\
    .groupby('category')\
    .agg(mean_cty = ('cty', 'mean'))
```

```python
# 문제 2) 앞 문제의 출력 결과는 `category` 값 알파벳순으로 정렬되어 있습니다.
# 어떤 차종의 도시 연비가 높은지 쉽게 알아볼 수 있도록 `cty(도시연비)` 평균이 높은 순으로 정렬해 출력하세요.
mpg\
    .groupby('category')\
    .agg(mean_cty = ('cty', 'mean'))\
    .sort_values('mean_cty', ascending = False)
```

```python
# 문제 3) 어떤 회사 자동차의 `hwy(고속도로 연비)`가 가장 높은지 알아보려고 합니다.
# `hwy(고속도로 연비)` 평균이 가장 높은 회사 세 곳을 출력하세요.

mpg\
    .groupby('manufacturer')\
    .agg(mean_hwy = ('hwy', 'mean'))\
    .sort_values('mean_hwy', ascending = False)\
    .head(3)
```

```python
# 문제 4) 어떤 회사에서 `compact` 차종을 가장 많이 생산하는지 알아보려고 합니다.
# 회사별 `compact` 차종 수를 내림차순으로 정렬해 출력하세요.

mpg\
    .query('category == "compact"')\
    .groupby('manufacturer')\
    .agg(total = ('manufacturer', 'count'))\
    .sort_values('total', ascending = False)
```

```python
mpg\
    .query('category == "compact"')\
    .value_counts('manufacturer')
```
