```python
import pandas as pd

exam = pd.read_csv('../sample-data/Doit_Python-main/Data/exam.csv');
exam.info()
```

```python
# 1-8) 1반이면서 수학 점수가 50점 이상인 경우
exam.query('nclass == 1 & math >= 50')
```

```python
# 1-9) 2반이면서 영어 점수가 80점 이상인 경우
exam.query('nclass == 2 & english >= 80')
```

```python
# 6-10) 수학 점수가 90점 이상이거나 영어 점수가 90점 이상인 경우
exam.query('math >= 90 | english >= 90')
```

```python
# 6-11) 영어 점수가 90점 미만이거나 과학 점수가 50점 미만인 경우
exam.query('english < 90 | science <= 50')
```

```python
# 6-12) 1, 3, 5반에 해당하는 케이스(row) 쿼리하기
exam.query('\
    nclass == 1 |\
    nclass == 3 |\
    nclass == 5\
')
```

```python
exam.query('nclass in [1, 3, 5]')
```

```python
# 6-13) 추출한 행으로 데이터 만들기
## 1, 2반 각각의 DataFrame 으로 추출하기
nclass1 = exam.query('nclass == 1')
nclass2 = exam.query('nclass == 2')
```

```python
# 6-14) 문자 변수를 이용해 조건에 맞는 행 추출하기
df = pd.DataFrame({
    'sex': ['F', 'M', 'F'],
    'country': ['Korea', 'USA', 'Japan'],
})
df
```

```python
# 6-15) 전체 조건에 작은따옴표, 추출할 문자에 큰따옴표 사용
df.query('sex == "F" & country == "Korea"')
```

```python
# 6-16) 전체 조건에 큰따옴표, 추출할 문자에 작은따옴표 사용
df.query('sex == "F" & country == "Japan"')
```

```python
# 6-17) DataFrame 외부의 변수를 query에 사용하는 방법: "@외부_변수명"
var = 3
exam.query('nclass == @var')
```

```python
exam.query(f'nclass == {var}')
```
