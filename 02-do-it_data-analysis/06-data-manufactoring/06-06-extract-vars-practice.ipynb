```python
import pandas as pd

mpg = pd.read_csv('../sample-data/Doit_Python-main/Data/mpg.csv')
mpg
```

```python
# 문제 1) mpg 데이터에서 `category(자동차 종류)`, `cty(도시 연비)` 변수를 추출해 새로운 데이터를 만드세요.
# 새로 만든 데이터의 일부를 출력해 두 변수로만 구성되어 있는지 확인하세요.
mpg_category_cty = mpg[['category', 'cty']]
mpg_category_cty
```

```python
suv_mean = mpg_category_cty\
    .query('category == "suv"')\
    ['cty']\
    .mean()
print('suv_mean: ', suv_mean)

compact_mean = mpg_category_cty\
    .query('category == "compact"')\
    ['cty']\
    .mean()
print('compact_mean: ', compact_mean)
```
