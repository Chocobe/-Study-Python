```python
import pandas as pd

mpg = pd.read_csv('../sample-data/Doit_Python-main/Data/mpg.csv');
mpg
```

```python
# 문제 1) 자동차 배기량에 따라 고속도로 연비가 다른지 알아보려고 합니다. `displ(배기량)`이 4 이하인 자동차와 5 이상인 자동차 중 어떤 자동차의 `hwy(고속도로 연비)` 평균이 더 높은지 알아보세요.
displ_low = mpg.query('displ <= 4')
displ_high = mpg.query('displ >= 5')

displ_low_hwy = displ_low['hwy'].mean()
displ_high_hwy = displ_high['hwy'].mean()

print('displ <= 4 의 고속도로 연비 평균: ', displ_low_hwy)
print('displ >= 5 의 고속도로 연비 평균: ', displ_high_hwy)
```

```python
# 문제 2) 자동차 제조 회사에 따라 도시 연비가 어떻게 다른지 알아보려고 합니다. `audi` 와 `toyota` 중 어느 `manufacturer(자동차 제조 회사)`의 `cty(도시 연비)` 평균이 더 높은지 알아보세요.
audi = mpg.query('manufacturer == "audi"')
toyota = mpg.query('manufacturer == "toyota"')

audi_cty_mean = audi['cty'].mean()
toyota_cty_mean = toyota['cty'].mean()

print('audy 도시 연비: ', audi_cty_mean)
print('toyota 도시 연비: ', toyota_cty_mean)
```

```python
# 문제 3) `chevrolet`, `ford`, `honda` 자동차의 고속도로 연비 평균을 알아보려고 합니다. 세 회사의 데이터를 추출한 다음 `hwy` 전체 평균을 구해 보세요.
chevrolet = mpg.query('manufacturer == "chevrolet"')
ford = mpg.query('manufacturer == "ford"')
honda = mpg.query('manufacturer == "honda"')

num_of_chevrolet_hwy = len(chevrolet['hwy'])
total_chevrolet_hwy = chevrolet['hwy'].sum()

num_of_ford_hwy = len(ford['hwy'])
total_ford_hwy = ford['hwy'].sum()

num_of_honda_hwy = len(honda['hwy'])
total_honda_hwy = honda['hwy'].sum()

total_mean = (total_chevrolet_hwy + total_ford_hwy + total_honda_hwy) / (
    num_of_chevrolet_hwy + num_of_ford_hwy + num_of_honda_hwy
)
total_mean
```
