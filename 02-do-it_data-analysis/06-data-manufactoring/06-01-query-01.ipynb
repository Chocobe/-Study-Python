```python
import pandas as pd

exam = pd.read_csv('../sample-data/Doit_Python-main/Data/exam.csv')
exam
```

```python
# 1) 조건에 맞는 데이터만 추출하기
nclass_counts = exam['nclass'].value_counts()
nclass_counts
```

```python
# 1-1) nclass 가 2인 케이스(row)만 쿼리하기
exam.query('nclass == 2')
```

```python
# 1-2) nclass 가 1이 아닌 케이스(row)만 쿼리하기
exam.query('nclass != 1')
```

```python
# 1-3) nclass 가 3이 아닌 케이스(row)만 쿼리하기
exam.query('nclass != 3')
```

```python
# 1-4) 수학 점수가 50점을 초과한 경우 쿼리하기
exam.query('math > 50')
```

```python
# 1-5) 수학 점수가 50점 미만인 경우 쿼리하기
exam.query('math < 50')
```

```python
# 1-6) 영어 점수가 50점 이상인 경우
exam.query('english >= 50')
```

```python
# 1-7) 영어 점수가 80점 이하인 경우
exam.query('english <= 80')
```
