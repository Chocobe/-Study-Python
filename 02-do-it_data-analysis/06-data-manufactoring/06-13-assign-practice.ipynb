```python
import pandas as pd

mpg = pd.read_csv('../sample-data/Doit_Python-main/Data/mpg.csv')
mpg
```

```python
# 문제 1) mpg 데이터 복사본을 만들고, `cty`와 `hwy`를 더한 `합산 연비 변수`를 추가하세요
mpg_clone = mpg.assign(
    total = lambda x: x['cty'] + x['hwy']
)
mpg_clone
```

```python
# 문제 2) 앞에서 만든 `합산 연비 변수`를 2로 나눠 `평균 연비 변수`를 추가하세요.
mpg_clone = mpg_clone.assign(
    mean = lambda x: x['total'] / 2
)
mpg_clone
```

```python
# 문제 3) `평균 연비 변수`가 가장 높은 자동자 3종의 데이터를 출력하세요.
mpg_clone\
    .sort_values(by = 'mean', ascending = False)\
    .head(3)
```

```python
# 문제 4) 1~3번 문제를 해결할 수 있는 하나로 연결된 pandas 구문을 만들어 실행해 보세요.
# 데이터는 복사본 대신 mpg 원본을 이용하세요.
mpg\
    .assign(
        total = lambda x: x['cty'] + x['hwy'],
        mean = lambda x: x['total'] / 2
    )\
    .sort_values(by = 'mean', ascending = False)\
    .head(3)
```
