```python
import pandas as pd

group_a = pd.DataFrame({
    'id': [1, 2, 3, 4, 5],
    'test': [60, 80, 70, 90, 85],
})
group_a
```

```python
group_b = pd.DataFrame({
    'id': [6, 7, 8, 9, 10],
    'test': [70, 83, 65, 95, 80],
})
group_b
```

```python
group_all = pd.concat([group_a, group_b])
group_all
```

```python
group_all_2 = pd.concat(
    [group_a, group_b],
    ignore_index = True
)
group_all_2
```
