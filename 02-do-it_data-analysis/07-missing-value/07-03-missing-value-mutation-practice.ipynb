```python
import pandas as pd
import numpy as np

mpg = pd.read_csv('/Users/young-woo/Desktop/study-2024/study-python/02-do-it_data-analysis/sample-data/Doit_Python-main/Data/mpg.csv')
mpg.loc[[64, 123, 130, 152, 211], ['hwy']] = np.nan
mpg
```

```python
# 문제 1) `drv(구동 방식)`별로 `hwy(고속도로 연비)` 평균이 어떻게 다른지 알아보려고 합니다.
# 분석을 하기전에 우선 두 변수에 결측치가 있는지 확인해야 합니다.
# `drv(구동 방식)` 변수와 `hwy(고속도로 연비)` 변수에 결측치가 몇 개 있는지 알아보세요.
drv_nan = mpg['drv'].isna().sum()
hwy_nan = mpg['hwy'].isna().sum()

print('drv_nan: ', drv_nan)
print('hwy_nan: ', hwy_nan)
```

```python
mpg[['drv', 'hwy']].isna().sum()
```

```python
# 문제 2) `df.dropna()`를 이용해 `hwy(고속도로 연비)` 변수의 결측치를 제거하고,
# 어떤 구동 방식의 `hwy(고속도로 연비)` 평균이 높은지 알아보세요.
# 하나의 `pandas` 구문으로 만들어야 합니다.

mpg\
    .dropna(subset = ['hwy'])\
    .groupby(['drv'])\
    .agg(hwy_mean = ('hwy', 'mean'))\
    .sort_values(by = ['hwy_mean'], ascending = [False])
```
