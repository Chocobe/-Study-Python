```python
# 문제 2) poptotal(전체 인구) 변수를 total로, popasian(아시아 인구) 변수를 asian으로 수정하세요
df_new = df.rename(columns = {
    'poptotal': 'total',
    'popasian': 'asian',
})
df_new
```

```python
# 문제 3) total, asian 변수를 이용해 '전체 인구 대비 아시아 인구 백분율' 파생변수를 추가하고, 히스토그램을 만들어 분포를 살펴보세요.
df_new['전체 인구 대비 아시아 인구 백분율'] = df_new['asian'] / df_new['total'] * 100
df_new
```

```python
df_new['전체 인구 대비 아시아 인구 백분율'].plot.hist()
```

```python
# 문제 4) 아시아 인구 백분율 전체 평균을 구하고, 평균을 초과하면 'large', 그 외에는 'small'을 부여한 파생변수를 만들어 보세요.
asian_mean = df_new['전체 인구 대비 아시아 인구 백분율'].mean()
asian_mean

# df_new['asian_size'] = np.where(df_new['전체 인구 대비 아시아 인구 백분율'] > asian_mean, 'large', 'small')
# df_new
```

```python
df_new['asian_size'] = np.where(df_new['전체 인구 대비 아시아 인구 백분율'] > asian_mean, 'large', 'small')
df_new
```

```python
# 문제 5) 'large'와 'small'에 해당하는 지역이 얼마나 많은지 빈도표와 빈도 막대 그래프를 만들어 확인해 보세요.
asian_size_counts = df_new['asian_size'].value_counts()
asian_size_counts
```
