```python
import pandas as pd

mpg = pd.read_csv('../sample-data/Doit_Python-main/Data/mpg.csv')
mpg
```

```python
# mpg 데이터에서 통합 연비 변수 만들기
mpg['total'] = (mpg['cty'] + mpg['hwy']) / 2
mpg
```

```python
# DataFrame 의 `mean()` 메서드로 평균 구하기
mpg['total'].mean()
```

```python
# 조건문을 활용해 파생변수 만들기
mpg.describe()
```

```python
# 값의 빈도수를 나타내는 그래프, `히스토그램(Histogram)` 렌더링 하기
# 히스토그램은 `도수분포표` 를 그래프로 나타낸 것이다.
mpg['total'].plot.hist()
```

```python
# 위에서 파악한 `total(연비)` 의 기준값을 `20` 으로 정하여,
# `고연비 합격 판정 (기준값 20 이상)` 을 만족하는 자동자 수 구하기
import numpy as np

mpg['test'] = np.where(mpg['total'] >= 20, 'pass', 'fail')
mpg
```

```python
# 빈도표로 합격 판정 자동차 수 살펴보기
# value_counts() 는 각 값의 빈도수를 도출해준다.
mpg['test'].value_counts()
```

```python
# `value_counts()` 를 사용하여 빈도수를 Bar(막대) 그래프로 렌더링하기
count_test = mpg['test'].value_counts()
count_test.plot.bar()
```

```python
# Bar(막대) 그래프의 X축 label 회전시키기
count_test.plot.bar(rot = 0)
```
